```python
# Functions - A Short Program: Zigzag

import time, sys
#-- Define function
"""
printZig: hàm vẽ 1 lượt của hình ZigZag
    heigh, width: chiều cao và chiều rộng của một lượt
    step: bước dịch chuyển
        nếu step > 0 : vẽ từ trái sang phải
        nếu step < 0: vẽ từ phải sang trái
        nếu step = 0: chỉ vẽ 1 line
"""
def printZig(height, width, step):
    global indent
    for i in range(height):
        print(' ' * indent, end = '')
        print('*' * width)
        indent += step
        time.sleep(0.1)

#-- Main Zigzag
indent = 5
zigzag_side = 5
step = 1
start_point = zigzag_side
try:
    while True:       
        printZig(zigzag_side, zigzag_side, step )      
        step *= -1 
except KeyboardInterrupt:
    sys.exit()
```

```python
#Lists - A Short Program: Conway’s Game of Life

# Define functions
def printTable(table):
    print('\n\n\n\n\n') # Separate each step with newlines.
    for i in table:
        for j in i:
            print(j, end = " ")         
        print()

# MAIN PROCESS   
import random, time, copy, sys
WIDTH = 8
HEIGHT = 8

# Create table:
nextTable = []
for i in range(HEIGHT):
    row = []
    for j in range(WIDTH):
        if random.randint(0,1) == 1:
            row.append("#") # add a alive cell
        else:
            row.append(" ") # add a dead cell
    nextTable.append(row)
    
try:
    while True:
        table = copy.deepcopy(nextTable)
        
        # Print table
        printTable(table)

        # Calculate the next step's cells based on current step's cells:
        for i in range(HEIGHT):
            for j in range(WIDTH):

                #Tính số sống sót xung quanh
                numAliveNeighbours = 0     
                for t in [-1,0,1]:
                    # xử lí vùng biên
                    if ((i + t) < 0) | ((i + t) >= HEIGHT ):
                        continue
                    for k in [-1,0,1]:
                        # xử lí vùng biên
                        if ((j + k) < 0) | ((j + k) >= WIDTH ):
                            continue
                        if not((t == 0) & (k == 0)):
                            if table[i+t][j+k] == "#":
                                numAliveNeighbours += 1

                #Tính sống/chết cho điểm(i,j)
                if (table[i][j] == "#"):
                    if (numAliveNeighbours not in [2,3]):
                        nextTable[i][j] = " "
                else:
                    if (numAliveNeighbours == 3):
                        nextTable[i][j] = "#"
        time.sleep(1) 
except KeyboardInterrupt:
    sys.exit()
```
